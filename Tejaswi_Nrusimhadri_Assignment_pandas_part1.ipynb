```python
import pandas as pd
df=pd.read_csv("/content/sample_data/orders.csv",sep=",")
df.set_index("order_id",inplace=True)
df.head(5)
df.tail(5)
```

```python
#Task 2 Run a structural inspection of df. From the output, answer the following in a markdown cell:

#How many rows and columns does the dataset have?
#Which columns have missing values and how many nulls does each have?
#Which column needs a data type conversion before any date-based analysis, and what function would you use?
```

```python
rows,cols=df.shape
print(f"number of rows are: {rows}\nNumber of columns are: {cols}\n")
#data=df.info()
#print(data)
missing_counts = df.isnull().sum()
missing_values=missing_counts[missing_counts>0]
#print(missing_values)
print("Missing values columns are :")
for index,value in missing_values.items():
  print(f"{index} has {value} null values")
for column in df.columns:
  if df[column].dtype=="object":
    try:
      pd.to_datetime(df[column], format="%Y-%m-%d")
      print(f"{column} Needs datetime conversion")
    except:
      pass
```

```python
import numpy as np

numeric_columns = df.select_dtypes(include=[np.number]).columns
print(f"Numeric columns are : {numeric_columns}")
median_price = df["unit_price"].median()
print("Median unit_price:", median_price)
#check unit_price is skewed
skew_val = df["unit_price"].skew()
print("Skewness of unit_price:", skew_val)
```
